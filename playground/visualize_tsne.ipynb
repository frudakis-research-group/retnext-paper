```python
import os
import glob

import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_paths = glob.glob('tsne_embeddings/augmentation_cubic_zero_clip*')

pretty_labels = {
    'CarbonDioxide_1_298K_0.05bar_mol/kg': 'CO$_2$ uptake at 298 K and 0.05 bar (mol/kg)',
    'CarbonDioxide_1_298K_0.5bar_mol/kg': 'CO$_2$ uptake at 298 K and 0.5 bar (mol/kg)',
    'CarbonDioxide_1_298K_2.5bar_mol/kg': 'CO$_2$ uptake at 298 K and 2.5 bar (mol/kg)',
    'Methane_1_298K_0.05bar_mol/kg': 'CH$_4$ uptake at 298 K and 0.05 bar (mol/kg)',
    'Methane_1_298K_0.5bar_mol/kg': 'CH$_4$ uptake at 298 K and 0.5 bar (mol/kg)',
    'Methane_1_298K_2.5bar_mol/kg': 'CH$_4$ uptake at 298 K and 2.5 bar (mol/kg)',
    'Hydrogen_1_77K_2bar_g/l': 'H$_2$ uptake at 77 K and 2 bar (g/l)',
    'Hydrogen_1_77K_100bar_g/l': 'H$_2$ uptake at 298 K and 100 bar (g/l)',
}
```

```python
csv_paths
```

```python
random_embeddings = pd.read_csv('tsne_embeddings/randomly_initialized_test.csv', index_col='name')
hMOF = pd.read_csv('/home/asarikas/databases/MOFXDB/hMOF/hMOF.csv', index_col='name')

#raise ValueError('VISUALIZE ONLY AFTER DECIDING ON THE CLIPPING SCHEME')
for path in csv_paths:
    #target = path.split('boltzmann_')[1].removesuffix('_test.csv')
    target = path.split('zero_clip_')[1].removesuffix('_test.csv')
    
    if 'Hydrogen' in target:
        units = 'g/l'
        target = target.replace('_g', f'_{units}')
    else:
        units = 'mol/kg'
        target = target.replace('mol', units)
        
    embeddings = pd.read_csv(path, index_col='name')
    
    fig, axes = plt.subplots(1, 2, figsize=[8, 4], layout='constrained', sharex=True, sharey=True)
    
    for ax, Z in zip(axes.flatten(), [random_embeddings, embeddings]):
        im = ax.scatter(
            Z.loc[:, 'tsne0'], Z.loc[:, 'tsne1'],
            c=hMOF.loc[Z.index, target],
            cmap='tab20c',
            s=1,
        )
    fig.colorbar(im, ax=ax, extend='max', label=pretty_labels[target])  # Common colorbar since they are colorized by same values.
    
    axes[0].set_xlabel('tSNE-1')
    axes[0].set_ylabel('tSNE-2')
    axes[1].set_xlabel('tSNE-1')
    
    axes[0].set_title('Before training')
    axes[1].set_title('After training')

    #fig.savefig(f'figures/tsne_{target.replace("/", "_")}.png', dpi=300)
```
